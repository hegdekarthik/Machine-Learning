```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv(r"C:\Users\khegde\Desktop\Salary_Data.csv")
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values

x_df = dataset.iloc[:, :-1]
y_df = dataset.iloc[:, 1]
```
